```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
```

```python
m = Chem.MolFromSmiles('CC1CN(c2cc3c(cc2F)c(=O)c(C(=O)O)cn3-c2ccc(F)cc2F)CCN1')
```

```python
m
```

```python
bit={}
morganfp=AllChem.GetMorganFingerprintAsBitVect(m,useChirality=True, radius=3, nBits = 2048, bitInfo=bit)

mfpvector = np.array(morganfp)
```

```python
np.nonzero(mfpvector)
```
